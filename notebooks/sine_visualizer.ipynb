# Summary

This notebook is used to visualize 2D datasets.

# Dataset

```python
from river.datasets import synth

dataset = synth.Sine(classification_function=3, balance_classes=True, has_noise=True)
```

# Execution

```python
x_list_0 = []
x_list_1 = []
y_list_0 = []
y_list_1 = []
label_list = []

for i, (x, y) in enumerate(dataset.take(2500)):
    if y:
        x_list_1.append(x[1])
        y_list_1.append(x[0])
    else:
        x_list_0.append(x[1])
        y_list_0.append(x[0])
```

# Plotting

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(x_list_0, y_list_0, ".", label="Label 0")
ax.plot(x_list_1, y_list_1, ".", label="Label 1")
ax.legend(loc="upper right")
ax.grid()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(x_list_0, y_list_0, ".", label="Labeled 0")
ax.plot(x_list_1, y_list_1, ".", label="Labeled 1")
ax.set_title("Function 1")
ax.legend()
ax.grid()
```
